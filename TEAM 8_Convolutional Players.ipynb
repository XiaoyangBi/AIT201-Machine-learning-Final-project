# 【1】Exploratory Data Analysis and visualize the counts of value in each columns

```python
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]] # For displaying purposes, pick columns that have between 1 and 50 unique values
    nRow, nCol = df.shape
    columnNames = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) / nGraphPerRow
    plt.figure(num = None, figsize = (6 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
```

# 【1】Dividing the data set into training and test sets (0.7 0.3)

```python
x_train = train_df.drop(['subject','Activity'], axis = 1)
y_train = train_df['Activity']
x_test = test_df.drop(['subject','Activity'], axis = 1)
y_test = test_df['Activity']
print("The x_train's shape is: ",x_train.shape)
print("The x_test's shape is:",x_test.shape)
```

# 【2】Data preprocessing

# Experiment 2：Test the accuracy of these three data pre-processing methods for the predicted data (x_train, x_test) in different machine learning models separately

```python
# Preprocessing the target column by LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_ = le.fit_transform(y_train)
y_test_ = le.fit_transform(y_test)
```

```python
# Preprocessing the data columns by StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_ = ss.fit_transform(x_train)
x_test_ = ss.fit_transform(x_test)
```

```python
# Preprocessing the data columns by MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_train__ = mms.fit_transform(x_train)
x_test__ = mms.fit_transform(x_test)
```

```python
# Preprocessing the data columns by Principal Component Analysis
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)

pca.fit(x_train)
pca.fit(x_test)

train_x_pca = pca.transform(x_train)
test_x_pca = pca.transform(x_test)

print(pca.n_components_)
print(pca.explained_variance_)
```

# 【3】Models, HyperparamterTuning

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import f1_score
```

```python
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()
knn = KNeighborsClassifier()
```

```python
# Using Standard Scaler
```

```python
# Fitting Random Forest To the training set 
# Fitting Decision Tree To the training set 
# Fitting K-Nearest Neighbor To the training set 

rfc.fit(x_train_, y_train_)
dtc.fit(x_train_, y_train_)
knn.fit(x_train_, y_train_)

y1_pred_ss = rfc.predict(x_test_)
y2_pred_ss = dtc.predict(x_test_)
y3_pred_ss = knn.predict(x_test_)

print('Model accuracy score of Random Forest Classifier with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test_, y1_pred_ss)))
print('Model accuracy score of Decision Tree Classifier with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test_, y2_pred_ss)))
print('Model accuracy score of KNeighbors Classifier with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test_, y3_pred_ss)))

print('Model f1 score of Random Forest Classifier with default hyperparameters: {0:0.4f}'. format(f1_score(y_test_, y1_pred_ss, average='weighted')))
print('Model f1 score score of Decision Tree Classifier with default hyperparameters: {0:0.4f}'. format(f1_score(y_test_, y2_pred_ss, average='weighted')))
print('Model f1 score score of KNeighbors Classifier with default hyperparameters: {0:0.4f}'. format(f1_score(y_test_, y3_pred_ss, average='weighted')))
```
